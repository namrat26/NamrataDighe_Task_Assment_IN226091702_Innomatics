### Problem Statement 1:Smart Parking Lot Management System
**Design a function to manage a smart parking lot.**\
The system should:
1. Accept vehicle entry and exit logs
2. Calculate total parked vehicles
3. Identify peak parking usage
4. Alert if parking exceeds capacity

**Real-Time Use**
- Mall parking systems
- Smart city infrastructure

**Hint**
- Function + list
- Loop for counting
- Conditional alert

**Sample Input**\
Parking Capacity: 50\
Vehicle Logs: ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

**Expected Output**\
Currently Parked Vehicles: 3\
Parking Status: Available

```python
def parking_lot_manager(capacity, vehicle_logs):
    current = 0
    peak = 0

    # Track vehicles using IN/OUT logs
    for log in vehicle_logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1  # prevents negative count

        # update peak usage
        if current > peak:
            peak = current

    # Capacity check
    if current > capacity:
        status = "Exceeded Capacity"
    elif current == capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)

parking_lot_manager(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

### Problem Statement 2:Online Food Delivery Time Estimator
**Create a function that estimates delivery time based on:**\
Distance (km)\
Weather condition\
Traffic level

Apply delays dynamically and display final ETA.\
**Real-Time Use**
- Food delivery apps
- Logistics platforms

**Hint**
- Function + multiple conditions
- Mathematical adjustments

**Sample Input**
- Distance (km): 8
- Traffic Level: High
- Weather: Rainy

**Expected Output**\
Estimated Delivery Time: 55 minutes


```python
def delivery_time_estimator(distance_km, traffic_level, weather):
    base_time = 5 + (distance_km * 5)

    # Traffic delay
    if traffic_level == "Low":
        traffic_delay = 0
    elif traffic_level == "Medium":
        traffic_delay = 10
    else:  # High
        traffic_delay = 20

    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 15
    else:  # Stormy
        weather_delay = 25

    eta = base_time + traffic_delay + weather_delay
    print("Estimated Delivery Time:", int(eta), "minutes")

# Sample Input
delivery_time_estimator(8, "High", "Rainy")
```

### Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, booked_seats_list):
    booked_count = len(booked_seats_list)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
        suggestion = "No additional show needed"
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider opening additional shows"
    else:
        status = "Available"
        suggestion = "No additional show needed"

    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)

# Sample Input (example: 150 bookings)
booked_seats = [1] * 150
seat_occupancy_analyzer(200, booked_seats)
```

### Problem Statement 4:Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    """
    Classifies server load based on average CPU usage:
    - < 50% -> Normal
    - 50% to 80% -> Warning
    - > 80% -> Critical
    """
    total = 0
    for value in cpu_readings:
        total += value

    avg_cpu = total / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg_cpu:.0f}%")
    print("Server Status:", status)

# Sample Input
server_load_classifier([45, 60, 70, 85, 90])
```

### Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def resource_usage_monitor(usage_hours):

    thresholds = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused = []

    for resource, hours in usage_hours.items():
        if resource in thresholds and hours > thresholds[resource]:
            overused.append(resource)

    # Print output
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Input
resource_usage_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

### Problem Statement 6: Online Event Registration Capacity Controller

```python
def registration_controller(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Sample Input
registration_controller(100, 105)
```
